```python
print("Hola Mundo")
```

```python
import numpy as np
import pandas as pd
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
diabetes_train = pd.read_csv('/content/diabetes_train.csv')
diabetes_train
```

```python
X = diabetes_train.iloc[:,0:8].values
X
```

```python
y = diabetes_train.iloc[:,8].values
y
```

## Definir la estructura de la Red Neuronal

```python
model = Sequential()
```

```python
model.add(Dense(12, input_dim=8, activation='relu'))
```

```python
model.add(Dense(9, activation='relu'))
```

```python
model.add(Dense(1, activation='sigmoid'))
```

## Modelo de compilaciÃ³n de la Red

```python
model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

## Entrenamiento de una red Neuronal

```python
historial = model.fit(X, y, epochs=1000, batch_size=10)
```

## Visualizar funciÃ³n de perdida

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel("# Epoca")
plt.ylabel("FunciÃ³n de pÃ©rdida")

plt.plot(historial.history['loss'])
```

## Test

```python
diabetes_test = pd.read_csv('/content/diabetes_test.csv')
diabetes_test
```

```python
X_test = diabetes_train.iloc[:,0:8].values
X_test
```

```python
y_test = diabetes_train.iloc[:,-1].values
y_test
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
np.round(y_pred)
```

```python
resultados = pd.DataFrame(np.round(y_pred), columns=['Prediccion'])
resultados['Real'] = pd.Series(y_test)
resultados
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
matrix = confusion_matrix(resultados['Real'], resultados['Prediccion'])
```

```python
disp = ConfusionMatrixDisplay(matrix)
disp.plot()
```
